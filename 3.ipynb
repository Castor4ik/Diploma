```python
#імпорт датасету
df= pd.read_csv("appstore_games.csv")
df
```

```python
#відкидання іменованих параметрів, що не мають сенсу для моделі

df = df.drop(columns = ['URL'], );
df = df.drop(columns = ['ID']);
df = df.drop(columns = ['Name']);
df = df.drop(columns = ['Subtitle']);
df = df.drop(columns = ['Icon URL'], );
df = df.drop(columns = ['Description'], );
df = df.drop(columns = ['Developer'], )
df = df.drop(columns = ['Primary Genre'], )
df = df.drop(columns = ['Genres'], )
df
```

```python
#перейменування змінних для коректної роботи з ними
df.rename(columns={'Average User Rating': 'Rating'}, inplace=True)
df.rename(columns={'User Rating Count': 'Count'}, inplace=True)
df.rename(columns={'In-app Purchases': 'Donates'}, inplace=True)
df.rename(columns={'Age Rating': 'Age'}, inplace=True)
df.rename(columns={'Original Release Date': 'Release'}, inplace=True)
df.rename(columns={'Current Version Release Date': 'CurrentRelease'}, inplace=True)
df
```

```python
#відкідаємо пропущені значення, тому що тут недоцільно використовувати заповнювачі через специфіку даних
df = df.dropna()
df
```

```python
#відкидаємо ігри, у яких менше за 50 відгуків, тому що там можуть бути накручені коментарі
df = df.loc[df['Count'] > 50]
```

```python
#відкідаємо пропущені значення ще раз
df = df.dropna()
df
```

```python
#для параметрів дати через те, що датасет не 2022 року, актуально брати тільки номер року (усі ігри, звісно, пізніше 2000)
df['Release'] = df['Release'].str[7:]
df['CurrentRelease']=df['CurrentRelease'].str[7:]
df['Release'] = df['Release'].astype(float, errors = 'raise')
df['CurrentRelease'] = df['CurrentRelease'].astype(float, errors = 'raise')
df
```

```python
#оскільки параметр останнього оновлення вже є неактуальним на зараз, то ми зробимо параметр, який враховує, коли було останнє оновлення відносно 2019 року
df['UpdateGap'] = 19 - df['CurrentRelease']
df = df.drop(columns = ['CurrentRelease'], )
df
```

```python
#для параметру донатів врахуємо середню ціну пропозиції у кожній грі
for i in df.index:
    my_list = df["Donates"][i].split(", ")
    my_list = [float(x) for x in my_list]
    df["Donates"][i] = np.mean(my_list)
    
df["Donates"] = df["Donates"].astype(float)
df.head()
```

```python
#створимо нову змінну для підрахунку кількості наявних локалізацій у грі
df['LanguagesQuantity'] = 0
df
```

```python
#підрахуємо кількість мов у кожній грі
import re
regex_pattern = re.compile(r", ")
for i in df.index:
    df["LanguagesQuantity"][i] = len(regex_pattern.findall(df["Languages"][i])) + 1
df
```

```python
#параметр вікових категорій - їх всього 4 і вони чітко розміщені за жорсткістю
df['Age'].unique()
```

```python
#замінимо параметр на числовий для моделювання
df['Age'] = df['Age'].str.replace('4+','0')
df['Age'] = df['Age'].str.replace('9+','1')
df['Age'] = df['Age'].str.replace('12+','2')
df['Age'] = df['Age'].str.replace('17+','3')
df['Age'] = df['Age'].str.replace('+','')
df['Age'] = df['Age'].astype(float, errors = 'raise')
df
```

```python
#видалимо вже непотрібний параметр
df = df.drop(columns = ['Languages'], )
```

```python
#міняємо порядок змінних для дослідження, оцінка стає крайнім правим стовпчиком
new_order = [1,2,3,4,5,6,7,8,0]
df = df[df.columns[new_order]]
df
```

```python
def plot_2d(X, label_list_cluster):
    plt.figure(figsize=(7, 7))
    plt.subplots_adjust(bottom=0.1)
    c = label_list_cluster
    plt.scatter(X[:, 0], X[:, 1], c=c, cmap='rainbow', label='True Position')
    plt.show()
```

```python
plot_2d(df[['Release', 'Count']].to_numpy(),hc4)
```

```python

# Function Max Leaf
def max_leaf_nodes(X_train, X_test, y_train, y_test, n):
    mse_train = []
    mse_test = []
    for i in n:
        ct = DecisionTreeClassifier(max_leaf_nodes = i, random_state=10).fit(X_train, y_train)
        mse_train.append(mean_squared_error(y_train, ct.predict(X_train)))
        mse_test.append(mean_squared_error(y_test, ct.predict(X_test)))
    
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(n, mse_train, alpha=0.5, color='blue', label='train')
    ax.plot(n, mse_test, alpha=0.5, color='red', label='test')
    ax.set_ylabel("MSE")
    ax.set_xlabel("max_leaf_nodes")
```

```python
# The optimal number of max_leaf_nodes
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
max_leaf_nodes(X_train, X_test, y_train, y_test, [2, 4, 6, 8, 10])
```
